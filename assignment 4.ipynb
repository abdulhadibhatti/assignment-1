1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance
is insufficient.

```python
class BankAccount:
    def __init__(self, account_holder_name, account_number, balance):
        self.account_holder_name = account_holder_name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}")
        print(f"New Balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
            print(f"Remaining Balance: {self.balance}")
        else:
            print("Insufficient Balance!")

account1 = BankAccount("Abdul Hadi", "123456789", 5000)

account1.deposit(2000)

account1.withdraw(3000)

account1.withdraw(5000)
```

2. Design a class 'Student' with attributes name and marks. Create a method to calculate grade
based on marks using conditions (A, B, C, Fail).

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 70:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "Fail"

student1 = Student("Abdul Hadi", 75)

print("Name:", student1.name)
print("Marks:", student1.marks)
print("Grade:", student1.calculate_grade())
```

3. Create a class 'ElectricityBill' for a KE-electric consumer. Include units consumed and calculate
bill based on slab rates (e.g., different price per unit for different ranges).

```python
class ElectricityBill:
    def __init__(self, consumer_name, units_consumed):
        self.consumer_name = consumer_name
        self.units_consumed = units_consumed

    def calculate_bill(self):
        if self.units_consumed <= 100:
            bill = self.units_consumed * 10
        elif self.units_consumed <= 200:
            bill = (100 * 10) + ((self.units_consumed - 100) * 15)
        else:
            bill = (100 * 10) + (100 * 15) + ((self.units_consumed - 200) * 20)

        return bill

    def display_bill(self):
        print("Consumer Name:", self.consumer_name)
        print("Units Consumed:", self.units_consumed)
        print("Total Bill: Rs.", self.calculate_bill())

consumer1 = ElectricityBill("Abdul Hadi", 250)

consumer1.display_bill()
```

5. Create a class 'Employee' with basic salary. Add a method to calculate total salary including
bonus (10%) and tax deduction (5%).

```python
class Employee:
    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = basic_salary

    def calculate_total_salary(self):
        bonus = self.basic_salary * 0.10      # 10% bonus
        tax = self.basic_salary * 0.05        # 5% tax deduction
        total_salary = self.basic_salary + bonus - tax
        return total_salary

emp1 = Employee("Abdul Hadi", 50000)

print("Employee Name:", emp1.name)
print("Basic Salary:", emp1.basic_salary)
print("Total Salary:", emp1.calculate_total_salary())
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# Circle Class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

circle = Circle(7)
rectangle = Rectangle(10, 5)

print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```

7. Create a class 'HospitalPatient' with attributes name, age, and disease. Add a method to
determine priority (Emergency if age > 60 or serious disease).

```python
class HospitalPatient:
    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease

    def determine_priority(self):
        serious_diseases = ["Heart Attack", "Cancer", "Stroke"]

        if self.age > 60 or self.disease in serious_diseases:
            return "Emergency"
        else:
            return "Normal"

patient1 = HospitalPatient("Abdul Hadi", 65, "Fever")

print("Name:", patient1.name)
print("Age:", patient1.age)
print("Disease:", patient1.disease)
print("Priority:", patient1.determine_priority())
```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item_name, price):
        self.items.append((item_name, price))

    def calculate_total(self):
        total = sum(price for item, price in self.items)
        return total

    def apply_discount(self):
        total = self.calculate_total()

        if total > 5000:  # Discount condition
            discount = total * 0.10  # 10% discount
            total -= discount
            print("Discount Applied: Rs.", discount)

        return total


cart = ShoppingCart()

cart.add_item("Laptop Bag", 2000)
cart.add_item("Mouse", 1500)
cart.add_item("Keyboard", 2500)

print("Total Bill Before Discount: Rs.", cart.calculate_total())
print("Final Bill After Discount: Rs.", cart.apply_discount())
```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Library:
    def __init__(self):
        self.books = ["Python", "Java", "C++", "DBMS"]

    def borrow_book(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"{book} has been issued.")
        else:
            print(f"Sorry! {book} is not available.")

    def return_book(self, book):
        self.books.append(book)
        print(f"{book} has been returned.")

    def display_books(self):
        print("Available Books:", self.books)


library = Library()

library.display_books()

library.borrow_book("Python")

library.display_books()

library.return_book("Python")

library.display_books()
```

```python
class Shape:
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


circle = Circle(5)
print("Area of Circle:", circle.area())

triangle = Triangle(10, 6)
print("Area of Triangle:", triangle.area())
```

```python
from abc import ABC, abstractmethod
import uuid

class Flight(ABC):
    def __init__(self, seats, price):
        self.seats = seats
        self.__ticket_price = price  # Encapsulation

    def get_price(self):
        return self.__ticket_price

    def set_price(self, price):
        self.__ticket_price = price

    @abstractmethod
    def flight_type(self):
        pass


class Economy(Flight):
    def flight_type(self):
        return "Economy"


class Business(Flight):
    def flight_type(self):
        return "Business"


class FirstClass(Flight):
    def flight_type(self):
        return "First Class"


class Passenger:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Booking:
    def __init__(self, passenger, flight):
        self.passenger = passenger
        self.flight = flight
        self.booking_id = str(uuid.uuid4())[:8]

    def calculate_bill(self):
        price = self.flight.get_price()

        if self.passenger.age < 12:
            price *= 0.50      # 50% discount for children
        elif self.passenger.age >= 60:
            price *= 0.80      # 20% discount for senior citizens

        tax = price * 0.12
        total = price + tax

        return total

    def make_payment(self, installments=False):
        total = self.calculate_bill()

        if installments:
            print(f"Payment in 3 Installments: {total/3:.2f} each")
        else:
            print(f"Full Payment: {total:.2f}")

    def book_seat(self):
        if self.flight.seats > 0:
            self.flight.seats -= 1
            print("Seat booked successfully.")
            return True
        else:
            print("No seats available.")
            return False

    def generate_ticket(self):
        total = self.calculate_bill()

        print("\n===== AIRLINE TICKET =====")
        print("Booking ID :", self.booking_id)
        print("Passenger  :", self.passenger.name)
        print("Age        :", self.passenger.age)
        print("Class      :", self.flight.flight_type())
        print("Total Fare :", round(total, 2))
        print("==========================")

        with open("booking_data.txt", "a") as file:
            file.write(
                f"{self.booking_id}, "
                f"{self.passenger.name}, "
                f"{self.passenger.age}, "
                f"{self.flight.flight_type()}, "
                f"{round(total,2)}\n"
            )


flight1 = Economy(50, 10000)

passenger1 = Passenger("Ali", 65)

booking1 = Booking(passenger1, flight1)

if booking1.book_seat():
    booking1.make_payment(installments=True)
    booking1.generate_ticket()
```

```python
from abc import ABC, abstractmethod
import uuid

class Product(ABC):
    def __init__(self, name, price, stock):
        self.name = name
        self.__price = price      
        self.stock = stock

    def get_price(self):
        return self.__price

    def set_price(self, price):
        self.__price = price

    @abstractmethod
    def product_type(self):
        pass


class AC(Product):
    def product_type(self):
        return "AC"


class Refrigerator(Product):
    def product_type(self):
        return "Refrigerator"


class WashingMachine(Product):
    def product_type(self):
        return "Washing Machine"


class Customer:
    def __init__(self, name):
        self.name = name
        self.cart = []

    def add_to_cart(self, product):
        if product.stock > 0:
            self.cart.append(product)
            product.stock -= 1
            print(f"{product.name} added to cart.")
        else:
            print(f"Sorry! {product.name} is out of stock.")


class Invoice:
    def __init__(self, customer):
        self.customer = customer
        self.invoice_id = str(uuid.uuid4())[:8]

    def calculate_total(self):
        total = sum(item.get_price() for item in self.customer.cart)

        if total > 100000:
            discount = total * 0.10   # 10% discount
            total -= discount
            print(f"Discount Applied: Rs.{discount}")

        tax = total * 0.18
        total += tax

        return total

    def payment(self, installment=False):
        total = self.calculate_total()

        if installment:
            print(f"3 Installments: Rs.{total/3:.2f} each")
        else:
            print(f"Full Payment: Rs.{total:.2f}")

        return total

    def generate_invoice(self):
        total = self.calculate_total()

        print("\n========== INVOICE ==========")
        print("Invoice ID :", self.invoice_id)
        print("Customer   :", self.customer.name)
        print("\nProducts:")

        for item in self.customer.cart:
            print(f"- {item.name} : Rs.{item.get_price()}")

        print("\nTotal Bill :", round(total, 2))
        print("=============================")

        with open("shop_records.txt", "a") as file:
            file.write(
                f"Invoice ID: {self.invoice_id}, "
                f"Customer: {self.customer.name}, "
                f"Total: {round(total,2)}\n"
            )


ac = AC("Haier AC", 80000, 5)
fridge = Refrigerator("Dawlance Refrigerator", 60000, 3)
wm = WashingMachine("LG Washing Machine", 50000, 2)

customer1 = Customer("Ali")

customer1.add_to_cart(ac)
customer1.add_to_cart(fridge)

invoice = Invoice(customer1)

invoice.payment(installment=True)
invoice.generate_invoice()
```

```python
from abc import ABC, abstractmethod
import uuid

class FoodItem(ABC):
    def __init__(self, name, price, quantity):
        self.name = name
        self.__price = price      # Encapsulation
        self.__quantity = quantity

    def get_price(self):
        return self.__price

    def get_quantity(self):
        return self.__quantity

    def reduce_quantity(self):
        if self.__quantity > 0:
            self.__quantity -= 1

    @abstractmethod
    def item_type(self):
        pass


class Pizza(FoodItem):
    def item_type(self):
        return "Pizza"


class Burger(FoodItem):
    def item_type(self):
        return "Burger"


class Drink(FoodItem):
    def item_type(self):
        return "Drink"


class Customer:
    def __init__(self, name):
        self.name = name
        self.cart = []

    def add_to_cart(self, item):
        if item.get_quantity() > 0:
            self.cart.append(item)
            item.reduce_quantity()
            print(f"{item.name} added to cart.")
        else:
            print(f"Sorry! {item.name} is unavailable.")


class Order:
    def __init__(self, customer):
        self.customer = customer
        self.order_id = str(uuid.uuid4())[:8]

    def calculate_total(self):
        total = sum(item.get_price() for item in self.customer.cart)

        if total > 2000:
            discount = total * 0.10
            total -= discount
            print(f"Discount Applied: Rs.{discount:.2f}")

        tax = total * 0.15
        total += tax

        return total

    def payment(self, method):
        total = self.calculate_total()

        print("\nPayment Method:", method)
        print("Amount Paid: Rs.", round(total, 2))

        return total

    def generate_invoice(self):
        total = self.calculate_total()

        print("\n========= FOOD INVOICE =========")
        print("Order ID :", self.order_id)
        print("Customer :", self.customer.name)

        print("\nOrdered Items:")
        for item in self.customer.cart:
            print(f"- {item.name} : Rs.{item.get_price()}")

        print("\nFinal Bill: Rs.", round(total, 2))
        print("================================")

        with open("food_orders.txt", "a") as file:
            file.write(
                f"Order ID: {self.order_id}, "
                f"Customer: {self.customer.name}, "
                f"Total Bill: Rs.{round(total,2)}\n"
            )


pizza = Pizza("Chicken Fajita Pizza", 1500, 5)
burger = Burger("Zinger Burger", 700, 10)
drink = Drink("Cold Drink", 300, 8)

customer = Customer("Ali")

customer.add_to_cart(pizza)
customer.add_to_cart(burger)
customer.add_to_cart(drink)

order = Order(customer)

order.payment("Card")

order.generate_invoice()
```
