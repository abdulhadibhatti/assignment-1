# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
result = int(input("Enter your number"))

if result >= 90:
    print("A")
elif result >= 80:
    print("B")
elif result >= 70:
    print("C")
elif result >= 60:
    print("D")
else:
    print("fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day_number = int(input("Enter day"))

if day_number == 1:
    print("It's Monday")
elif day_number == 2:
    print("It's Tuesday")
elif day_number == 3:
    print("It's Wednesday")
elif day_number == 4:
    print("It's Thursday")
elif day_number == 5:
    print("It's Friday")
elif day_number == 6:
    print("It's Saturday")
elif day_number == 7:
    print("It's Sunday")
else:
    print("week close")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month_number = input("Enter month_number ")

if month_number == [12, 1, 2]:
    print("Winter")
elif month_number == [3, 4, 5]:
    print("Sprint")
elif month_number == [6, 7, 8]:
    print("Summer")
else:
    print("Autumn")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print("Result:", a + b)

elif op == "-":
    print("Result:", a - b)

elif op == "*":
    print("Result:", a * b)

elif op == "/":
    print("Result:", a / b)

else:
    print("Invalid operator")
```
